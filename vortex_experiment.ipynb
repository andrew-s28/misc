```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
X = np.linspace(1,10,1000)
para = [10*x**2 for x in X]
```

```python
plt.plot(X,para)
```

```python
logX = np.log(X)
logpara = np.log(para)
slope,inter = st.linregress(logX,logpara)[0:2]
```

```python
slope,inter
```

```python
f = np.array([1.428571429,1.67,2.22222222,2.857142857])
H = np.array([2.5,5,9,14])
logf = np.log(f)
logH = np.log(H)
slope,inter = st.linregress(logf,logH)[0:2]
fig,axs = plt.subplots(1,2,constrained_layout=True)
axs[0].plot(logf,logH,'.')
axs[0].plot(logf,logf*slope+inter,label='Slope=')
axs[0].set_xlabel('log(f)')
axs[0].set_ylabel('log(H)')
print(slope,inter)
axs[1].plot(f,H,'.')
axs[1].plot(np.linspace(0,3,100),[np.exp(inter)*x**slope for x in np.linspace(0,3,100)])
axs[1].set_xlabel('f (Hz)')
axs[1].set_ylabel('H (mm)')
np.exp(inter)
```
